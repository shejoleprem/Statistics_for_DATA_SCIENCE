### Q2. Write a Python function to estimate the population mean using a sample mean and standard deviation.

```python
import math
def population_mean(sample_mean,sample_sd,sample_size):
    standard_error=sample_sd/math.sqrt(sample_size)
    z=1.96 #assume confidence interval is 95%
    lower_ci=sample_mean-z*standard_error
    upper_ci=sample_mean+z*standard_error
    population_mean=(lower_ci+upper_ci)/2
    return population_mean
```

```python
population_mean(4,1.5,50)
```

### Q.5 Write a Python script to conduct a hypothesis test on the difference between two population means, given a sample from each population.

```python
import numpy as np
from scipy import stats
np.random.seed(123)
sample1=np.random.normal(loc=10,scale=2,size=100)
sample2=np.random.normal(loc=10,scale=2,size=100)

mean1=np.mean(sample1)
mean2=np.mean(sample2)
std1=np.std(sample1,ddof=1)
std2=np.std(sample2,ddof=1)

t_stat, p_value=stats.ttest_ind(sample1,sample2)

print(f"sample 1 mean:{mean1:.2f},sample 2 mean:{mean1:.2f}")
print(f"sample sd:{std1:.2f},sample sd:{std2:.2f}")
print(f"t-statistics:{t_stat:.2f},p-value:{p_value:.3f}")

if p_value<0.05:
    print("The difference between the population means is statistically significant")
else:
    print("The difference between the population means is not statistically significant")
    
```

### Q9. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom parameter set to 10.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

df=10

x = np.linspace(t.ppf(0.01, df), t.ppf(0.99, df))

tdist=t(df)

y=tdist.pdf(x)

plt.plot(x,y,'r-',lw=2, label='t-distribution')

plt.xlabel("t-values")
plt.ylabel("probability density")

plt.show()
```

### Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two random samples of equal size and a null hypothesis that the population means are equal.

```python
import numpy as np
from scipy import stats

sample1=np.random.normal(loc=10,scale=2,size=100)
sample2=np.random.normal(loc=11,scale=2,size=100)

t_test,p_value=stats.ttest_ind(sample1,sample2)

print("t_test value {} and p_value {}".format(t_test,p_value))
```
